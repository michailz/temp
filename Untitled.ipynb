```python
import numpy as np
import pandas as pd
pd.options.display.max_columns = None
import json
import glob
```

## Load json files to dataframe df

```python
dfs = []
for file in glob.glob('*.json'):
    data = pd.read_json(file) # read data frame from json file
    dfs.append(data) # append the data frame to the list
df = pd.concat(dfs, ignore_index=True)
```

## We can see that message id is unique only within the Channel

```python
df.sort_values(by='id').head()
```

## Extract channel_id from peer_id

```python
df['channel_id'] = [d.get('channel_id') for d in df.peer_id]
df = df.drop(columns=['peer_id'])
```

## Extract from_id from dataframe

```python
column_name = 'from_id'
df_temp = df.loc[df[column_name].notna(), ['id', 'channel_id', column_name]]
df_temp2 = pd.DataFrame(df_temp[column_name].to_list())
df_temp2 = df_temp2.add_prefix(column_name + '_')
df_temp = pd.concat([df_temp.reset_index(drop=True), df_temp2.reset_index(drop=True)], axis=1)
df_temp = df_temp.drop(columns=[column_name])
df_temp.from_id_user_id = df_temp.from_id_user_id.astype('Int64')
df_temp.from_id_channel_id = df_temp.from_id_channel_id.astype('Int64')
from_id = df_temp
from_id
```
